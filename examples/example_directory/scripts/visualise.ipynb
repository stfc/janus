# Visualising Training

To plot the output of of the networks training, functions for plotting two styles of graph are available.

```python
%matplotlib inline
import cc_hdnnp.visualisation as vis
```

## Learning Curve

A plot of training and test errors for either energy or force against the epoch can be generated from the directory that training was performed in (specifically, it must contain the `learning-curve.out` file). By default, the energy will be plotted:

```python
n2p2_directory = "../../m_cresol/all_frames/n2p2_1"
vis.plot_learning_curve(n2p2_directory=n2p2_directory)
```

Alternatively, any of the headers present in `learning-curve.out` can be used in the plot with the `keys` argument:

```python
vis.plot_learning_curve(n2p2_directory=n2p2_directory, keys=["RMSE_Ftrain_pu", "RMSE_Ftest_pu"])
```

## Epoch Accuracy

For a more detailed look at the performance at a single epoch, plots of reference energy/force versus the value predicted by the network can be used:

```python
vis.plot_epoch(n2p2_directory=n2p2_directory, epoch=5)
```

Note that because there are far more values for the force, to speed up plotting the number of points is subsampled by default. The degree to which this occurs can be controlled with the `subsample_forces` argument.

```python
vis.plot_epoch(n2p2_directory=n2p2_directory, epoch=10, subsample_forces=100)
```
